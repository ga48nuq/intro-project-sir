```python
def SEIR(y_0,t,beta,gamma,sigma,mu,muI):
    """ODEs of the SEIR model"""
    dSdt = mu - (beta * y_0[2] + mu) * y_0[0]
    dEdt = beta * y_0[2] * y_0[0] - (mu + sigma) * y_0[1]
    if mortal: 
        dIdt = sigma * y_0[1] - (mu + muI + gamma) * y_0[2]
    else: 
        dIdt = sigma * y_0[1] - (mu + gamma) * y_0[2]
    dRdt = gamma * y_0[2] - mu * y_0[3]
    
    return [dSdt, dEdt, dIdt, dRdt]

# Change beta parameter due to seasons 
def season_beta(beta, beta_1, t):
    return beta * (1 + beta_1 * np.cos(2 * np.pi * (t/365)))

# SEIR equations with seasonal influence
def SEIRseas(y_0,t,beta,gamma,sigma,mu,muI,beta1):
    """ODEs of the SEIR model incorporating seasonality"""
    
    b_seas = season_beta(beta, beta_1, t) # Calculate beta parameter due to influence
    dSdt = mu - (b_seas * y_0[2] + mu) * y_0[0]
    dEdt = b_seas * y_0[2] * y_0[0] - (mu + sigma) * y_0[1]
    
    if mortal: #Add mortality parameter
        dIdt = sigma * y_0[1] - (mu + muI + gamma) * y_0[2]
    else: 
        dIdt = sigma * y_0[1] - (mu + gamma) * y_0[2]
        
    dRdt = gamma * y_0[2] - mu * y_0[3]
    
    return [dSdt, dEdt, dIdt, dRdt] 

mortal = True
seasons = True

beta,beta1,gamma,mu,sigma, muI = 2.5,0.5,0.5,1/70,0.9,0.5

print('gamma {}, mu {}'.format(gamma, mu))
if mortal: 
    print('mu', muI)   
if seasons: #Vary beta parameter with seasons 
    beta_low = season_beta(beta, beta1, 298/2)
    beta_high = season_beta(beta, beta1, 0)
    print(beta_low, '\t<  Beta\t< ', beta_high)
else: 
    print('Beta\t= ', beta)
    R0 = beta/(gamma+mu)
if seasons: 
    print(round(beta_low/(gamma+mu),3), '<R0< ', round(beta_high/(gamma+mu),3))
else: print('R0\t= ', R0)

# Intial conditions 
I_0 = 0.01
E0 = 0
R0 = 0.00
S_0 = 1
y_0 = [S_0, E0, I_0, R0]

time = np.linspace(0,298,298) # 2 year time span

#  Fit ODE function
if seasons: 
    run = odeint(SEIRseas,y_0,time,args=(beta,gamma,sigma,mu,muI,beta1,))
else: 
    run = odeint(SEIR,y_0,time,args=(beta,gamma,sigma,mu,muI,))


    
################

# # Times at which the solution is to be computed
# t = np.linspace(timepoint1,timepoint2,timepoint2-timepoint1)

# # Create a Data object using our initiated data from above.
# data = Data(t, prop) 

# # Define initial conditions
# I_0 = 1/17280000
# R_0 = 0.0
# S_0 = 1
# y_0 = [S_0, I_0, R_0]

# # Store information of solveODEs function 
# model = Model(solveODEs) 
# # # Set up ODR with the model, data and initial parameter estimate.
# odr = ODR(data, model, beta0=beta_guess) 
# # # Run the regression
# out = odr.run() 
# beta,gamma=out.beta

# print("beta gamma: {} found after fitting".format(out.beta))
# print("Standard deviations found after fitting: {}".format(out.sd_beta))

# #Plot proportion infected over time 
# plt.figure(figsize=(960/my_dpi, 480/my_dpi), dpi=my_dpi)
# # plt.plot(t, prop,'ro',color='tab:green',markersize=2)
# plt.plot(t, solveODEs((beta,gamma),t),color='tab:blue',markersize=10)
# plt.plot(t, run,color='tab:blue',markersize=10)
# plt.xlabel('Time (days)')
# plt.ylabel('Proportion Infected')
# plt.grid(True)
# plt.vlines(lockdown,0,max_val,color='black',linestyle='dashed')
# #     plt.vlines(lockdown2,0,max_val,color='black',linestyle='dashed')
# # plt.savefig('model_fit.png')
# plt.show()


################
    

plt.plot(time,run)
# plt.legend(('Susceptibles','Exposed','Infected','Recovered'),loc='upper right')
# plt.xlabel('time (days)')
# plt.ylabel('proportion')
# plt.grid(True)
# plt.savefig('seir_s1_m0.png')
# plt.show()

# #Vary initial parameters for I and S to obtain phase plot
# initials = []
# for i in range(7):
#     I_0 = 0.01 + 0.15 * i
#     E0 = 0
#     S_0 = 1 - E0 - I_0 - R0
#     y_0 = [S_0, E0, I_0, R0]
#     initials.append(odeint(SEIR,y_0,time,args=(beta,gamma,sigma,mu,muI,)))
#     plt.plot(initials[i][:,0],initials[i][:,2],color='blue')
#     plt.scatter(initials[i][0,0],initials[i][0,2],color='blue')
# initials = []
# for i in range(5):
#     I_0 = 0.01
#     E0 = 0
#     S_0 = 1 - I_0 - (0.1 * i)
#     R0 = 1 - E0 - I_0 - S_0
#     y_0 = [S_0, E0, I_0, R0]
#     initials.append(odeint(SEIR,y_0,time,args=(beta,gamma,sigma,mu,muI,)))
#     plt.plot(initials[i][:,0],initials[i][:,2],color='blue')
#     plt.grid(True)
#     plt.scatter(initials[i][0,0],initials[i][0,2],color='blue')

# #Phase plot
# plt.xlabel('proportion Susceptible')
# plt.ylim(0,1)
# plt.grid(True)
# plt.ylabel('proportion Infected')
# plt.plot([1.0, 0.01], [0.0, 1.0], linestyle=':',color='blue')
# plt.savefig('seir_phase_s1_m0.png')
# plt.show()

```

```python
def SIR(y,t,beta,gamma):
    '''
    Ordinary Differential Equations of the basic SIR model
    '''
    S0,I0,R0 = y
    dSdt = - beta * S0 * I0
    dIdt = beta * S0 * I0 - gamma * I0
    dRdt = gamma * I0   
    return [dSdt, dIdt, dRdt] 


def get_cases(country=nl,citizens=17280000):
    '''
    Return proportional cases of infectious individuals 
    '''
    # Convert experimental numbers to proportions
    number = [i for i in country['cases']]
    # prop = [i/(17*10**6) for i in number]
    prop = [i/citizens for i in number]

    return prop


def plot_cases_sir(timepoint1=0,timepoint2=150,beta_guess=[1, 0.8],prop=propnl,population=pop_nl,
                  lockdown=75, max_val=0.0005):

        
    def solveODEs(param,t):
        """Solves differential equations"""
        beta,gamma = param

         # define differential equations
        def SIR(y_0, t):
            """The Ordinary Differential Equations for the basic SIR model"""
            dSdt = - beta * y_0[0] * y_0[1]
            dIdt = beta * y_0[0] * y_0[1] - gamma * y_0[1]
            dRdt = gamma * y_0[1]
            return [dSdt, dIdt, dRdt] 

        return odeint(SIR, y_0, t)[:,1]

    # Times at which the solution is to be computed
    t = np.linspace(timepoint1,timepoint2,timepoint2-timepoint1)

    # Create a Data object using our initiated data from above.
    data = Data(t, prop) 

    # Define initial conditions
    I_0 = 1/17280000
    R_0 = 0.0
    S_0 = 1
    y_0 = [S_0, I_0, R_0]

    # Store information of solveODEs function 
    model = Model(solveODEs) 
    # # Set up ODR with the model, data and initial parameter estimate.
    odr = ODR(data, model, beta0=beta_guess) 
    # # Run the regression
    out = odr.run() 
    beta,gamma=out.beta

    print("beta gamma: {} found after fitting".format(out.beta))
    print("Standard deviations found after fitting: {}".format(out.sd_beta))

    #Plot proportion infected over time 
    plt.figure(figsize=(960/my_dpi, 480/my_dpi), dpi=my_dpi)
    plt.plot(t, prop,'ro',color='tab:green',markersize=2)
    plt.plot(t, solveODEs((beta,gamma),t),color='tab:blue',markersize=10)
    plt.xlabel('Time (days)')
    plt.ylabel('Proportion Infected')
    plt.grid(True)
    plt.vlines(lockdown,0,max_val,color='black',linestyle='dashed')
#     plt.vlines(lockdown2,0,max_val,color='black',linestyle='dashed')
    # plt.savefig('model_fit.png')
    plt.show()

# Plot for Germany separetely
propgerm = get_cases(germ)
plot_cases_sir(prop=propgerm[:150],population=pop_germ)
plot_cases_sir(timepoint1=150,timepoint2=len(germ),
               beta_guess=[.09, .05],prop=propgerm[150:len(germ)],
               population=pop_germ,max_val=0.0006)


# plot_cases_sir(prop=propnl,timepoint2=298,beta_guess=[0.9, 0.8])


# Plot for the Netherlands separetely
# propnl = get_cases()
# plot_cases_sir(prop=propnl[:150])
# plot_cases_sir(timepoint1=150,timepoint2=len(nl),beta_guess=[.09, .05],prop=propnl[150:len(nl)],
#               max_val=0.0008,lockdown=288)
```

```python
def go_out(country, out_df=out_df):
    """
    Return the estimated number of days people usually go out.
    @country <str>: coutry name
    @out_df <pd.DataFrame>: template dataframe
    """
    df = out_df.copy()
    try:
        series = df[country]
    except KeyError:
        raise KeyError(f"Population pyramid data of {country} is not defined!")
    df = df.iloc[:, :6]
    df["Portion"] = series
    return df
```
